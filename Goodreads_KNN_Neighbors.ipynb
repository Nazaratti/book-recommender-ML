# KNN Nearest Neighbors

```python
from scipy.sparse import csr_matrix

from sklearn.neighbors import NearestNeighbors

import recmetrics
```

```python
us_rating_pivot = final_df.head(200000).pivot(index = 'Book_Title', columns = 'User_ID', values = 'Book_Rating_x').fillna(0)
us_rating_matrix = csr_matrix(us_rating_pivot.values)
```

```python
us_rating_pivot.head()
```

```python
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'auto')
model_knn.fit(us_rating_matrix);
```

```python
us_rating_pivot[us_rating_pivot.index.str.contains('')]
```

### Code to randomly select a title for recommendation

```python
model_knn.effective_metric_
```

```python
query_index = np.random.choice(us_rating_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(us_rating_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 11)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(us_rating_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of: {2}'.format(i, us_rating_pivot.index[indices.flatten()[i]], round(distances.flatten()[i],2)))
```

### Generate recommendations for a list of titles

```python
books_list = ['Crime and Punishment', 'The Analyst', 'The Pillars of the Earth', 'Anne Frank: The Diary of a Young Girl']
rec_number = int(input('Number of Recommendations: '))

for book_name in books_list:
    distances, indices = model_knn.kneighbors(us_rating_pivot[us_rating_pivot.index.str.contains(book_name)].iloc[0].values.reshape(1,-1), n_neighbors = rec_number + 1)
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(us_rating_pivot[us_rating_pivot.index.str.contains(book_name)].iloc[0].name))
        else:
            print('{0}: {1}, with distance of: {2}'.format(i, us_rating_pivot.index[indices.flatten()[i]], round(distances.flatten()[i],2)))
    print('\n')
```
